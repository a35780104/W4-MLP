```python
def load_img(f):
    f=open(f)
    lines=f.readlines()
    
    imgs,lab=[], []
    
    for i in range(len(lines)):
        fn,label = lines[i].split(' ')
        im1 = cv2.imread(fn)
        print (i,type(im1))
        im1 = cv2.resize(im1,(50,50))
        im1 = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)
        

        imgs.append(im1)
        lab.append(int(label))
        
    return imgs, lab
```

```python
import cv2
x_train,y_train = load_img('train.txt')
x_val,y_val = load_img('val.txt')
```

```python
import numpy as np
x_train=np.array(x_train)
y_train=np.array(y_train)
x_val=np.array(x_val)
y_val=np.array(y_val)
```

```python
print( 'train feature datas =', y_train.shape )
```

```python
train_feature_vector = x_train.reshape( len(x_train), 50*50 ).astype('float32')
train_feature_vector /= 255
```

```python
from keras.utils import np_utils

train_label_onehot = np_utils.to_categorical(y_train)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras import Sequential

model = Sequential()

model.add( Dense( input_dim = 50*50             
                , units = 256                   
                , kernel_initializer = 'normal' 
                , activation = 'relu'           
                )
         )

model.add( Dense( units = 50                    
                , kernel_initializer = 'normal' 
                , activation = 'softmax'        
                )
         )
```

```python
model.compile( loss='categorical_crossentropy' 
             , optimizer = 'adam'              
             , metrics = ['accuracy']          
             )


history = model.fit(               
          x = train_feature_vector 
        , y = train_label_onehot   
        , validation_split = 0.2   
        , epochs = 75              
        , batch_size = 1000        
        , verbose = 2              
)
```

```python
import matplotlib.pyplot as plot  

def train_history_graphic( history       
                         , history_key1 
                         , history_key2  
                         , y_label       
                         ) :
    
    plot.plot( history.history[history_key1] )


    plot.plot( history.history[history_key2] )


    plot.title( 'train history' )


    plot.xlabel( 'epochs' )


    plot.ylabel( y_label )

    plot.legend( ['train', 'validate']
               , loc = 'upper left'
               )

    plot.show()

train_history_graphic( history, 'accuracy', 'val_accuracy', 'accuracy' )
```

```python
test_feature_vector = x_val.reshape( len(x_val), 50*50 ).astype('float32')
test_feature_vector /= 255

test_label_onehot = np_utils.to_categorical(y_val)

eval = model.evaluate( test_feature_vector, test_label_onehot )
print( 'loss =', eval[0] )
print( 'val_accuracy =', eval[1] )
```

```python
x_test,y_test = load_img('test.txt')
x_test = np.array(x_test)
y_test = np.array(y_test)
```

```python
test_feature_vector = x_test.reshape( len(x_test), 150*150 ).astype('float32')
test_feature_vector /= 255

test_label_onehot = np_utils.to_categorical(y_test)

eval = model.evaluate( test_feature_vector, test_label_onehot )

print( 'loss =', eval[0] )
print( 'test_accuracy =', eval[1] )
```
